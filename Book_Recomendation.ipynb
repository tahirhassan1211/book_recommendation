### Popularity Based Recommendation

```python
# Create column average ratings using groupby on book title because some of the books have same book title but different isbn number.

book_rating_user['Avg_Ratings'] =  book_rating_user.groupby('title')['book_rating'].transform('mean')

# Create column number of ratings
book_rating_user['No_Of_Ratings'] = book_rating_user.groupby('title')['book_rating'].transform('count')

# Create new data frame popular_df
df= book_rating_user[['title','Avg_Ratings','No_Of_Ratings']]
```

```python
# Drop duplicate books

df.drop_duplicates('title',inplace=True)
```

```python
# Applying a constraint of number of ratings greater tha 200 to only keep the most popular books 

df =df[df['No_Of_Ratings']>200].sort_values('Avg_Ratings',ascending=False)
```

### Content Based Recommendation

```python
Content_based =book_rating_user[["title","author","publisher","Country"]]
Content_based
```

```python
# Removing all the spaces between the words and combing them into a letter
def remove(string):
    ns=""
    for i in string:
        if(not i.isspace()):
            ns+=i
    return ns   
Content_based["author"] = Content_based["author"].apply(remove)
Content_based["publisher"]   = Content_based["publisher"].apply(remove)
```

```python
Content_based["tags"] = Content_based["author"] + "," + Content_based["publisher"] + "," + Content_based["Country"]
```

```python
Content_based
```

```python
Content_based.drop(["author","publisher","Country"],axis=1,inplace=True)
```

```python
data=Content_based.head(5000)
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000)
```

```python
vector = (cv.fit_transform(data["tags"])).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_content = cosine_similarity(vector)
```

```python
def recommend_content(book_name):
    index = data[data["title"]== book_name].index[0]
    distances = sorted(list(enumerate(similarity_content[index])),reverse=True,key = lambda x: x[1])
    books_recommended =[]
    for i in distances[1:6]:
         books_recommended.append(data.iloc[i[0]]["title"])
    return books_recommended   
```

```python
recommend_content("Timeline")
```
